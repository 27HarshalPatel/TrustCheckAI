```python
### Sample Jupyter Notebook: Data Loading, Preprocessing, and Fairness Metrics with AIF360

# Install AIF360 if not already installed
# !pip install aif360

import warnings
warnings.filterwarnings("ignore")

# Import core libraries
import numpy as np
import pandas as pd

# Import AIF360 datasets and metrics
from aif360.datasets import AdultDataset, CompasDataset
from aif360.metrics import BinaryLabelDatasetMetric, ClassificationMetric

# ------------------------------
# 1. Load Datasets
# ------------------------------
print("Loading Adult Income dataset...")
adult_data = AdultDataset()

print("Loading COMPAS dataset...")
compas_data = CompasDataset()

# ------------------------------
# 2. Define Protected Attributes
# ------------------------------
# Adult Income: Protected attributes are 'sex' and 'race'
privileged_groups_adult = [{'sex': 1}]   # Male
unprivileged_groups_adult = [{'sex': 0}] # Female

# COMPAS: Protected attribute is 'race'
privileged_groups_compas = [{'race': 1}]   # Caucasian
unprivileged_groups_compas = [{'race': 0}] # African-American

# ------------------------------
# 3. Fairness Metrics - Adult Dataset
# ------------------------------
print("\nFairness Metrics for Adult Income Dataset:")
adult_metric = BinaryLabelDatasetMetric(
    adult_data,
    unprivileged_groups=unprivileged_groups_adult,
    privileged_groups=privileged_groups_adult
)

print("Statistical Parity Difference:", adult_metric.statistical_parity_difference())
print("Disparate Impact:", adult_metric.disparate_impact())

# ------------------------------
# 4. Fairness Metrics - COMPAS Dataset
# ------------------------------
print("\nFairness Metrics for COMPAS Dataset:")
compas_metric = BinaryLabelDatasetMetric(
    compas_data,
    unprivileged_groups=unprivileged_groups_compas,
    privileged_groups=privileged_groups_compas
)

print("Statistical Parity Difference:", compas_metric.statistical_parity_difference())
print("Disparate Impact:", compas_metric.disparate_impact())

# ------------------------------
# 5. Summary
# ------------------------------
print("\nNOTE:")
print("- Statistical Parity Difference close to 0 means fair outcomes across groups.")
print("- Disparate Impact close to 1 means fair outcomes across groups.")
```
